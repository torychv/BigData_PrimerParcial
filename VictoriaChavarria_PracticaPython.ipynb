# Ejercicio 1: Clasificación de edad  

**Enunciado:** determinar si una persona es niño, adolescente, adulto o adulto mayor según su edad.  

**Lógica:** se evalúa la edad con condiciones if, elif y else. El primer rango que se cumple determina la categoría.

```python
def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"
edad = 21
categoria = clasificar_edad(edad)
print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")
```

# Ejercicio 2: Conversor de temperatura  

**Enunciado:** convertir una temperatura de grados Celsius a Fahrenheit.  

**Lógica:** se utiliza la fórmula F = (C × 9/5) + 32 para convertir la temperatura.  

```python
def convertir_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
celsius = 30
resultado = convertir_fahrenheit(celsius)
print(f"Temperatura en Celsius: {celsius}°C")
print(f"Temperatura en Fahrenheit: {resultado}°F")
```

# Ejercicio 3: Calculadora básica  

**Enunciado:** realizar operaciones matemáticas básicas entre dos números.  

**Lógica:** se utilizan condiciones para identificar la operación seleccionada y validar que no se divida entre cero.

```python
def calculadora(num1, num2, operacion):
    if operacion == "+":
        return num1 + num2
    elif operacion == "-":
        return num1 - num2
    elif operacion == "*":
        return num1 * num2
    elif operacion == "/":
        if num2 == 0:
            return "Error: no se puede dividir entre cero"
        return num1 / num2
    else:
        return "Operación no válida"
num1 = 20
num2 = 5
operacion = "/"
resultado = calculadora(num1, num2, operacion)
print(f"Número 1: {num1}")
print(f"Número 2: {num2}")
print(f"Operación: {operacion}")
print(f"Resultado: {resultado}")
```

# Ejercicio 4: Número par o impar

**Enunciado:** determinar si un número entero es par o impar.

**Lógica:** se utiliza el operador módulo (%) para obtener el residuo de la división entre 2. Si el residuo es igual a 0, el número es par; de lo contrario, es impar.

```python
def verificar_numero(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Impar"

numero = 15
resultado = verificar_numero(numero)

print(f"Número: {numero}")
print(f"Resultado: {resultado}")
```

# Ejercicio 5: Tabla de multiplicar

**Enunciado:** mostrar la tabla de multiplicar de un número entero desde 1 hasta 12.

**Lógica:** se utiliza un ciclo for para recorrer los números del 1 al 12. En cada iteración se multiplica el número definido por el valor del ciclo y se muestra el resultado.

```python
numero = 7

print(f"Tabla de multiplicar del {numero}")

for i in range(1, 13):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

# Ejercicio 6: Contador de vocales

**Enunciado:** contar cuántas vocales contiene una frase.

**Lógica:** se recorre cada carácter de la frase y se verifica si pertenece al conjunto de vocales en mayúsculas o minúsculas. Cada vez que se encuentra una vocal, el contador aumenta en uno.

```python
frase = "Python es un lenguaje muy util"

contador = 0
vocales = "aeiouAEIOU"

for letra in frase:
    if letra in vocales:
        contador += 1

print(f"Frase: {frase}")
print(f"Cantidad de vocales: {contador}")
```

# Ejercicio 7: Promedio de calificaciones

**Enunciado:** calcular el promedio de 5 calificaciones e indicar si el estudiante aprobó o reprobó.

**Lógica:** se suman las 5 calificaciones, se calcula el promedio y se evalúa con una condición `if` si el promedio es mayor o igual a 70.

```python
def calcular_promedio(calificaciones):
    suma = sum(calificaciones)
    promedio = suma / len(calificaciones)

    if promedio >= 70:
        resultado = "Aprobó"
    else:
        resultado = "Reprobó"

    return promedio, resultado


calificaciones = [85, 70, 90, 60, 75]

promedio, resultado = calcular_promedio(calificaciones)

print(f"Calificaciones: {calificaciones}")
print(f"Promedio: {promedio}")
print(f"Resultado: {resultado}")
```

# Ejercicio 8: Lista de números positivos y negativos

**Enunciado:** contar cuántos números son positivos, negativos y ceros dentro de una lista de 10 números.

**Lógica:** se recorre la lista usando un ciclo `for` y se evalúa cada número con condiciones `if`, `elif` y `else`.

```python
def contar_numeros(lista):
    positivos = 0
    negativos = 0
    ceros = 0

    for numero in lista:
        if numero > 0:
            positivos += 1
        elif numero < 0:
            negativos += 1
        else:
            ceros += 1

    return positivos, negativos, ceros


numeros = [5, -3, 0, 8, -1, 0, 12, -7, 4, -2]

positivos, negativos, ceros = contar_numeros(numeros)

print(f"Números: {numeros}")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Ceros: {ceros}")
```

# Ejercicio 9: Mayor y menor de una lista

**Enunciado:** encontrar el número mayor y menor de una lista y calcular la diferencia entre ambos.

**Lógica:** se utilizan las funciones `max()` y `min()` para encontrar el valor mayor y menor de la lista. Luego se calcula la diferencia restando ambos valores.

```python
def analizar_lista(lista):
    mayor = max(lista)
    menor = min(lista)
    diferencia = mayor - menor

    return mayor, menor, diferencia


numeros = [15, 8, 23, 4, 42, 16, 9]

mayor, menor, diferencia = analizar_lista(numeros)

print(f"Lista: {numeros}")
print(f"Número mayor: {mayor}")
print(f"Número menor: {menor}")
print(f"Diferencia: {diferencia}")
```

## Explicación de `max()` y `min()`

- `max()` devuelve el número más grande de una lista.
- `min()` devuelve el número más pequeño de una lista.

**Lógica:** se utiliza un diccionario para guardar los datos del estudiante en pares clave-valor. Luego se recorren los datos con un ciclo `for` para mostrarlos de forma ordenada.

# Ejercicio 11: Función de descuento

**Enunciado:** calcular el precio final con descuento.

**Lógica:** se crea una función que recibe el precio y el descuento, y retorna el resultado aplicando la fórmula.

```python
def calcular_descuento(precio, descuento):
    return precio - (precio * descuento / 100)

print(calcular_descuento(100, 20))
```

# Ejercicio 12: Validación de contraseña

**Enunciado:** verificar si una contraseña es válida.

**Lógica:** se evalúan condiciones como longitud mínima, presencia de mayúsculas, minúsculas y números usando funciones como `any()`.

```python
password = "Abc1234567"

valida = (
    len(password) >= 8 and
    any(c.isupper() for c in password) and
    any(c.islower() for c in password) and
    any(c.isdigit() for c in password)
)

print("Contraseña valida" if valida else "Contraseña invalida")
```

# Ejercicio 13: Sistema de inventario

**Enunciado:** gestionar una lista de productos.

**Lógica:** se usa una lista de diccionarios para almacenar productos y se recorren para mostrar y calcular el valor total.

```python
productos = [
    {"codigo": 1, "nombre": "Laptop", "precio": 1000, "cantidad": 5},
    {"codigo": 2, "nombre": "Mouse", "precio": 20, "cantidad": 10}
]

for p in productos:
    print(p)

total = sum(p["precio"] * p["cantidad"] for p in productos)
print("Valor total:", total)
```

# Ejercicio 14: Registro de ventas

**Enunciado:** crear una lista de ventas que contenga cliente, producto, cantidad y precio unitario. Calcular el total por venta, el total general vendido e identificar el cliente con la compra más alta.

**Lógica:** se utiliza una lista de diccionarios para almacenar la información de cada venta. Se recorre la lista con un ciclo para calcular el total de cada venta y acumular el total general. También se compara cada venta para encontrar la compra con el valor más alto.

```python
ventas = [
    {"cliente": "Juan", "producto": "Laptop", "cantidad": 1, "precio_unitario": 15000},
    {"cliente": "Maria", "producto": "Mouse", "cantidad": 2, "precio_unitario": 300},
    {"cliente": "Carlos", "producto": "Monitor", "cantidad": 1, "precio_unitario": 4500}
]

total_general = 0
compra_mas_alta = 0
cliente_mayor_compra = ""

print("Registro de ventas")

for venta in ventas:
    total_venta = venta["cantidad"] * venta["precio_unitario"]
    total_general += total_venta

    print(f"Cliente: {venta['cliente']}")
    print(f"Producto: {venta['producto']}")
    print(f"Total venta: L. {total_venta}")
    print()

    if total_venta > compra_mas_alta:
        compra_mas_alta = total_venta
        cliente_mayor_compra = venta["cliente"]

print(f"Total general vendido: L. {total_general}")
print(f"Cliente con la compra más alta: {cliente_mayor_compra}")
print(f"Monto de la compra más alta: L. {compra_mas_alta}")
```

# Ejercicio 15: Limpieza básica de texto

**Enunciado:** limpiar una frase eliminando espacios al inicio y al final, convertirla a minúsculas, reemplazar espacios dobles por uno solo y contar cuántas palabras contiene.

**Lógica:** se utilizan métodos de cadenas para limpiar y transformar el texto. `strip()` elimina espacios innecesarios al inicio y final, `lower()` convierte todo a minúsculas, `replace()` sustituye espacios dobles por uno simple y `split()` divide la frase en palabras para poder contarlas.

```python
frase = "   Python  es un lenguaje MUY util para Big Data   "

texto_limpio = frase.strip()
texto_limpio = texto_limpio.lower()
texto_limpio = texto_limpio.replace("  ", " ")

palabras = texto_limpio.split()

print(f"Texto original: '{frase}'")
print(f"Texto limpio: '{texto_limpio}'")
print(f"Cantidad de palabras: {len(palabras)}")
```

## Explicación de los métodos utilizados

- **strip()**: elimina los espacios en blanco que se encuentran al inicio y al final de una cadena de texto.

- **lower()**: convierte todos los caracteres del texto a letras minúsculas.

- **replace()**: permite reemplazar una cadena de caracteres por otra dentro del texto. En este ejercicio se utiliza para sustituir espacios dobles por un solo espacio.

- **split()**: divide una cadena en una lista de palabras utilizando los espacios como separador. Esto permite contar cuántas palabras contiene el texto.
